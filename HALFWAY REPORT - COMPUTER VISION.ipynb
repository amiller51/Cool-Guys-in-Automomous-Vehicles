```python
# IMPORT PYTHON PACKAGES
import cv2
import numpy as np


# DEFINE REGION OF INTEREST
def region_of_interest(img, vertices):
    mask = np.zeros_like(img)
    match_mask_color = 255
    cv2.fillPoly(mask, vertices, match_mask_color)
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


# CREATE BLANK IMAGE
# ADD IN LANE DETECTION LINE PROPERTIES
def drow_the_lines(img, lines):
    img = np.copy(img)
    blank_image = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)

    for line in lines:
        for x1, y1, x2, y2 in line:
            cv2.line(blank_image, (x1,y1), (x2,y2), (255, 255, 0), thickness=5)

    img = cv2.addWeighted(img, 2, blank_image, 1, 0.0)
    return img


# CREATE PROCESS IMAGE
# DEFINE VERTICES FOR MASKING
def process(image):
    print(image.shape)
    height = image.shape[0]
    width = image.shape[1]
    region_of_interest_vertices = [
        (1, height),
        (700, 450),
        (1000, height)
    ]
    
# CREATE DETECTION FILTERS 
# ADD LANE LINES TO PROCESSED IMAGE
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    canny_image = cv2.Canny(gray_image, 80, 100)
    cropped_image = region_of_interest(canny_image,
                    np.array([region_of_interest_vertices], np.int32),)
    lines = cv2.HoughLinesP(cropped_image,
                            rho=2,
                            theta=np.pi/180,
                            threshold=110,
                            lines=np.array([]),
                            minLineLength=25,
                            maxLineGap=80)
    image_with_lines = drow_the_lines(image, lines)
    return image_with_lines

# IMPORT VIDEO FILE
# APPLY COMPUTER VISION DETECTION FILTERS
# DISPLAY OUTPUT VIDEO WITH LANE LINES DETECTED
cap = cv2.VideoCapture('LaneDetection.mp4')

width= int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height= int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
writer = cv2.VideoWriter('lanedetvideo.mp4', cv2.VideoWriter_fourcc(*'MJPG'), 20, (width,height))

while cap.isOpened():
    ret, frame = cap.read()
    lanelines = process(frame)
    writer.write(frame)
    cv2.imshow('Lane Detection Video', lanelines)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
