```python
image = cv2.imread(r'C:\Users\Collin\Desktop\Auto Vehicles\roadSC.jpg')
```

```python
plt.imshow(image)
```

```python
b = image[:, :, 0]
g = image[:, :, 1]
r = image[:, :, 2]
```

```python
img = cv2.merge([r, g, b])
```

```python
plt.imshow(img)
```

```python
def show_with_matplotlib(color_img, title, pos):
    RGB = color_img[:, :, ::-1]
    
    ax = plt.subplot(1, 4, pos)
    plt.imshow(RGB)
    plt.axis('off')
```

```python
plt.figure(figsize=(20, 8))
```

```python
filtered = cv2.GaussianBlur(img, (3, 3), 0)
```

```python
gray = cv2.cvtColor(filtered, cv2.COLOR_BGR2GRAY)
```

```python
x = cv2.Sobel(gray, cv2.CV_16S, 0, 1, 3)
```

```python
absX = cv2.convertScaleAbs(x)
```

```python
show_with_matplotlib(img, "original", 1)
show_with_matplotlib(cv2.cvtColor(absX, cv2.COLOR_GRAY2BGR), "new", 2)
plt.figure(figsize=(20, 8))
```

```python
show_with_matplotlib(cv2.cvtColor(absX, cv2.COLOR_GRAY2BGR), "new", 2)
```
